## Wprowadzenie teoretyczne

### Ramy (Frames)
Ramy to struktury reprezentujące wiedzę na temat powtarzalnych sytuacji lub konceptów. W ramie określone są podstawowe elementy (tzw. sloty), które charakteryzują daną sytuację, oraz ich wartości.

### Skrypty (Scripts)
Skrypty to szczególny rodzaj ram, który opisuje typowy ciąg zdarzeń, czyli sekwencję kroków, które zazwyczaj mają miejsce w danej sytuacji.
